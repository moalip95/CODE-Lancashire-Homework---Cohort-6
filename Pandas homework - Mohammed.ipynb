```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(25,10))

df_pokemon['Type 1'].value_counts().plot(kind='pie', ax=axes[0])
df_pokemon['Type 2'].value_counts().plot(kind='pie', ax=axes[1])

axes[0].set_ylabel('Type 1')
axes[1].set_ylabel('Type 2')
```

```python
df_pokemon.plot(kind = 'scatter', x = 'Type 1', y = 'Attack')

plt.show()
```

```python
df_pokemon['Defense'].plot(kind = 'hist')
plt.grid(color='red', linestyle=':', linewidth=1.0, axis='y', alpha=0.5)

plt.ylabel('Defense')  # labling y-axis
plt.xlabel('Defense')           # labling x-axis
plt.title('An array of defense in Pokemon')

plt.show()
```
