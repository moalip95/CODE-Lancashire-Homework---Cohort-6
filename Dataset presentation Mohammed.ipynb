```python
#converting education expenditure and avg income from dollars to GBP
df_iq['avg_income_GBP'] = df_iq['avg_income']*0.82
df_iq['education_expenditure_GBP'] = df_iq['education_expenditure']*0.82
```

```python
df_iq.head()
```

```python
#removing original US dollars income and expenditure
df_iq.drop(['education_expenditure','avg_income'], axis=1)
```

```python
df_iq = df_iq.drop(['education_expenditure','avg_income'], axis=1)
df_iq.head()
```

```python
#finding what percentage of avg income is spent on education in each country
df_iq['exp_spend'] = df_iq['avg_income_GBP']/df_iq['education_expenditure_GBP']
df_iq['exp_spend']
df_iq.head()
```

```python
#rounding numbers to make data easier to view
df_iq = df_iq.round()
df_iq.head()
```

```python
#removing any country completely where one of the columns has no data available
df_iq.dropna(subset=['rank','country','IQ','avg_income_GBP','education_expenditure_GBP','exp_spend'], inplace=True)
df_iq.head()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(14, 8))

# Width parameter for bars. Decrease for more space between bars.
bar_width = 0.6

# First graph: Country and IQ
axs[0].bar(df_iq['country'], df_iq['IQ'], color='green')
axs[0].set_title('Country and IQ')
axs[0].set_xticks(range(len(df_iq['country'])))  # explicitly set x-ticks position
axs[0].set_xticklabels(df_iq['country'], rotation=90, fontsize=10)  # adjust fontsize if necessary

# Second graph: Country and education_expenditure
axs[1].bar(df_iq['country'], df_iq['education_expenditure_GBP'], color='red')
axs[1].set_title('Country and education_expenditure')
axs[1].set_xticks(range(len(df_iq['country'])))  # explicitly set x-ticks position
axs[1].set_xticklabels(df_iq['country'], rotation=90, fontsize=10)  # adjust fontsize if necessary

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(15, 8))
# Plotting average income on the primary y-axis
ax1.scatter(df_iq['IQ'], df_iq['avg_income_GBP'], color='green', marker='x', alpha=0.7, label='Average Income (GBP)')
ax1.set_xlabel('IQ', fontsize=14)
ax1.set_ylabel('Average Income (GBP)', fontsize=14, color='green')
ax1.tick_params(axis='y', labelcolor='green')
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)

# Creating a secondary y-axis for education expenditure
ax2 = ax1.twinx()
ax2.scatter(df_iq['IQ'], df_iq['education_expenditure_GBP'], color='blue', marker='o', alpha=0.7, label='Education Expenditure (GBP)')
ax2.set_ylabel('Education Expenditure (GBP)', fontsize=14, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')


fig.suptitle('Scatter Plot showing IQ against Education Expenditure and Average Income', fontsize=16)
fig.tight_layout()
plt.show()

#Generally positive correlation between income and IQ across all countries
```

```python
plt.scatter(df_iq['IQ'], df_iq['exp_spend'], color='red', alpha=0.7, edgecolor='black', s=50)
plt.xlabel('IQ')
plt.ylabel('Percentage of Income on expenditure')
plt.title('Relationship between IQ and percentage of average income spent on expenditure')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=12)
plt.show()
```

```python
fig = px.bar(df_iq.sort_values(by="rank"), x="rank", y="avg_temp", 
    title="Rank compared to Average Temperature",
    color="avg_temp", 
    color_continuous_scale=["blue","yellow","orange","red"],  # color gradient
    labels={"rank": "Country Rank", "avg_temp": "Average Temperature (°C)"})
fig.show()
#no correlation from my perspective
```

```python
plt.figure(figsize=(40, 50))

#Palette mapped to the 'rank'
sns.barplot(x='rank', y='country', data=df_iq, 
    palette="plasma", 
    linewidth=2.5, edgecolor=".2")

# Label axes and setting title
plt.xlabel('Rank', fontsize=18)
plt.ylabel('Country', fontsize=18)
plt.title('Rank of Country vs IQ', fontsize=22)

# Setting font size for the axis ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Displaying grid lines
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()





```

```python
#I want to show longer bars at the top so that the highest ranked country has a longer bar

df_iq['inverted_rank'] = df_iq['rank'].max() - df_iq['rank'] + 1
#Using the above, the max rank is reduced by the rank+1, which inverts the rank,
#which whill reverse the order of ranks, countries with lower rank numbers which 
#were originally higher IQs now have higher values to indicate the rank is now
#corresponding to IQ

plt.figure(figsize=(40, 50))

# Palette mapped to the 'inverted_rank'
sns.barplot(x='inverted_rank', y='country', data=df_iq, 
    palette="Spectral", 
    linewidth=2.5, edgecolor=".2")

# Label axes and setting title
plt.xlabel('Inverted Rank', fontsize=18)
plt.ylabel('Country', fontsize=18)
plt.title('Rank of Country vs IQ (Higher Bars Indicate Higher IQ)', fontsize=22)

# Setting font size for the axis ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Displaying grid lines
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()

```

```python
#Creating a pie chart using IQ but in ranges for countries

# Create IQ brackets
bins = [60, 70, 80, 90, 100, 110]
labels = ["60-69", "70-79", "80-89", "90-99", "100-109"]
df_iq['IQ_bracket'] = pd.cut(df_iq['IQ'], bins=bins, labels=labels, right=False)

# Count countries in each bracket
iq_counts = df_iq['IQ_bracket'].value_counts()

# Plotting
plt.figure(figsize=(10, 8))
plt.pie(iq_counts, labels=iq_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Countries by IQ Bracket')
plt.show()
```

```python
fig = px.bar(df_iq.sort_values(by="rank"), x="country", y=["IQ", "avg_income_GBP", "education_expenditure_GBP"], 
    title="Comparison of Key Metrics by Country Rank",
    labels={"country": "Country", "value": "Amount", "variable": "Metrics"},
    color_discrete_sequence=px.colors.qualitative.Set2,
    height=5*len(df_iq),  # Adjusting height based on number of countries
    width=20*len(df_iq)) #adjusting height and width for readability
#Above, I have sorted the data based on rank, to analyse avg income, expenditure and IQ
#The labels rename the axes, and legend

fig.update_layout(barmode='stack',
        xaxis_title="Country",
        yaxis_title="Value",
        legend_title="Metrics",
        xaxis_tickangle=-45,   # Rotating tick labels for better readability
        xaxis_tickfont_size=10,  # Adjusting font size of x-axis labels
        margin=dict(b=150)) #increasing bottom margin


fig.show()
```
